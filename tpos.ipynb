```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import statsmodels.api as sm 
import sklearn 
%matplotlib inline
```

```python
# csv inlezen met ',' als scheidingsteken
world_happiness = pd.read_csv("world-happiness-report.csv",delimiter=",")
world_happiness.head()
```

```python
world_happiness.info()
```

```python
# csv inlezen met ',' als scheidingsteken
world_happiness_2021 = pd.read_csv("world-happiness-report-2021.csv",delimiter=",")
world_happiness_2021.drop(world_happiness_2021.columns[[12,13,14,15,16,17,18,19]], axis=1, inplace=True)
world_happiness_2021.head()
```

```python
world_happiness_2021.info()
```

```python
world_happiness["Country name"].unique()
```

```python
# 20 happiest countries

plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
plt.title("20 happiest countries", fontsize=16)
plt.ylabel("Country name", fontsize=16)

barplot1 = world_happiness_2021.groupby("Country name")["Ladder score"].mean().sort_values(ascending=False).head(20)
sns.barplot(x=barplot1.values,y=barplot1.index)
plt.show()
```

```python
# 20 least happy countries

plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
plt.title("20 least happy countries", fontsize=16)
plt.ylabel("Country name", fontsize=16)

barplot1 = world_happiness_2021.groupby("Country name")["Ladder score"].mean().sort_values(ascending=False).tail(20)
sns.barplot(x=barplot1.values,y=barplot1.index)
plt.show()
```

```python
# test

plt.figure(figsize=(20,15))
sns.set_style("whitegrid")

world_happiness_2021.sort_values(by="Ladder score", inplace=True)
plt.title("lalala", fontsize=16)
plt.xlabel("Ladder score", fontsize=16)
plt.ylabel("Logged GDP per capita", fontsize=16)


sns.scatterplot(data=world_happiness_2021, y="Logged GDP per capita", x="Ladder score")

for i in range(len(world_happiness_2021)):
    plt.text(s = world_happiness_2021.loc[i,"Country name"], x=world_happiness_2021.loc[i,"Ladder score"]+0.01,y=world_happiness_2021.loc[i,"Logged GDP per capita"]+0.01, fontsize=10)

# for i in range(len(world_happiness_2021)):
    # plt.text(s = world_happiness_2021.loc[i,'Country name'], x=world_happiness_2021.loc[i,"Ladder score"]+0.01,y=world_happiness_2021.loc[i,"Logged GDP per capita"]+0.01, fontsize=10)

plt.show()
```
